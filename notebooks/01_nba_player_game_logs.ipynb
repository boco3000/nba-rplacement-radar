```python
nba_players = players.find_players_by_full_name("Stephen Curry")
nba_players
```

```python
player_id = nba_players[0]["id"]
player_id
```

```python
logs = playergamelogs.PlayerGameLogs(
    season_nullable="2024-25",
    season_type_nullable="Regular Season"
)

df_logs = logs.get_data_frames()[0]
df_player = df_logs[df_logs["PLAYER_ID"] == player_id].copy()

len(df_player)
```

```python
df_player["GAME_DATE"] = pd.to_datetime(df_player["GAME_DATE"])
df_player = df_player.sort_values("GAME_DATE").reset_index(drop=True)

df_player.head()
```

```python
type(df_player)
```

```python
len(df_player)
```

```python
season_df = df_player.copy()
recent_df = df_player.tail(5)

season_df.shape, recent_df.shape
```

```python
stats_cols = ['MIN', 'PTS', 'REB', 'AST']

season_avg = season_df[stats_cols].mean()
recent_avg = recent_df[stats_cols].mean()

season_avg, recent_avg
```

```python
delta = recent_avg - season_avg
delta
```

```python
summary = pd.DataFrame({
    "season_avg": season_avg,
    "recent_avg_last5": recent_avg,
    "delta_last5_minus_season": delta
})

summary
```

```python
butler = players.find_players_by_full_name("Jimmy Butler")
butler
```

```python
butler_id = butler[0]["id"]
butler_id
```

```python
butler_games = df_logs[df_logs["PLAYER_ID"] == butler_id].copy()

butler_games["GAME_DATE"] = pd.to_datetime(butler_games["GAME_DATE"])
butler_game_dates = set(butler_games["GAME_DATE"])

len(butler_game_dates)
```

```python
with_butler = df_player[df_player["GAME_DATE"].isin(butler_game_dates)]
without_butler = df_player[~df_player["GAME_DATE"].isin(butler_game_dates)]

len(with_butler), len(without_butler)
```

```python
stats_cols = ["MIN", "PTS", "REB", "AST"]

with_avg = with_butler[stats_cols].mean()
without_avg = without_butler[stats_cols].mean()

with_avg, without_avg
```

```python
impact_delta = without_avg - with_avg
impact_delta
```

```python
impact_summary = pd.DataFrame({
    "with_butler_avg": with_avg,
    "without_butler_avg": without_avg,
    "delta_without_minus_with": impact_delta
})

impact_summary
```
